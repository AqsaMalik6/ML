```python
#Loading Data and Showing Features
from sklearn.datasets import load_wine

wine = load_wine()
print(wine.keys())
print(wine["target_names"])

n_samples, n_features = wine.data.shape
print('Number of samples:', n_samples)
print('Number of features:', n_features)

print('Feature Names:', wine.feature_names)  
print('Target Names:', wine.target_names)  
print("Dimension of Input:", wine.data.shape) 
print("Dimension of Output:", wine.target.shape) 
print("First 5 rows of Input:", wine.data[0:5])  
print("Target Data:", wine.target)

```

```python
#HISTOGRAMS OF THE FEATURES
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
x_index = 3
colors = ['blue', 'red', 'green']
for label, color in zip(range(len(wine.target_names)), colors):
    ax.hist(wine.data[wine.target==label, x_index], label=wine.target_names[label], color=color)
    ax.set_xlabel(wine.feature_names[x_index])
    ax.legend(loc='upper right')
fig.show()
```

```python
#SCATTERPLOT WITH TWO FEATURES
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine

wine = load_wine()
fig, ax = plt.subplots()
x_index = 3
y_index = 0
colors = ['blue', 'red', 'green']
for label, color in zip(range(len(wine.target_names)), colors):
    ax.scatter(wine.data[wine.target == label, x_index], wine.data[wine.target == label, y_index], label=wine.target_names[label], c=color)
ax.set_xlabel(wine.feature_names[x_index])
ax.set_ylabel(wine.feature_names[y_index])
ax.legend(loc='upper right')  

plt.show()

```

```python
#Scatter Plot Total
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine

wine = load_wine()
n = len(wine.feature_names)
fig, ax = plt.subplots(n, n, figsize=(16, 16))
colors = ['blue', 'red', 'green']
for x in range(n):
    for y in range(n):
        xname = wine.feature_names[x]
        yname = wine.feature_names[y]
        for color_ind in range(len(wine.target_names)):
            ax[x, y].scatter(wine.data[wine.target==color_ind, x], wine.data[wine.target==color_ind, y], label=wine.target_names[color_ind], c=colors[color_ind])
        ax[x, y].set_xlabel(xname)
        ax[x, y].set_ylabel(yname)
        ax[x, y].legend(loc='upper left')

plt.show()

```

```python
#Convert into Pandas
import pandas as pd
wine_df = pd.DataFrame(wine.data, columns=wine.feature_names)
wine_df['Target']=wine.target
print(wine_df.head(5))
print(wine_df.info())
wine_df['Target'] = wine_df['Target'].astype(object)
print(wine_df.info())
print(wine_df.groupby('Target').size())
```
